```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stat
import seaborn as sns
import pandas as pd
```

we would like to compute the posterior distribution of the following model:

x_1...x_n | mu, tau distributed as N(mu, tau^-1)

mu distributed as N(mu_o, (k_0 *tau*)^-1)

tau distributed as Gamma(a_0, b_0)



Now imagine we want to approximate the posterior with the product of two functions p(z|x) = q_(mu)* q_(tau)
and then we will confront this with the real posterior distribution that is analitically know 



```python
# we'll suppose data coming from N(5,1)
N = 1000
x_real = np.random.normal(5,1, N)

mu_0 = 3
a_0 = 2
b_0 = 3
k_0= 2

mu_n = (k_0*mu_0 + np.sum(x_real))/(k_0+N) # known 
a_n = a_0 + (N+1)/2 #known 

#initialize tau_n to some arbitrary value 
tau_n = 1 
control= True
n_iter = 0


fig, axs = plt.subplots(ncols=3, nrows=2, figsize = (15,10))
######################## TRUE PROCEDIMENTO ################

mean= np.mean(x_real)
tau_true_sample = np.random.gamma(a_0+N/2, np.power(b_0 + 1/2 * np.sum(np.power(x_real-mean,2)) + k_0*N/(2*(k_0+N))*(mean-mu_0)**2,-1), size= 10000)
mu_true = np.sum(x_real)/(N+k_0)+ k_0 /(k_0+N)*mu_0
mu_t_sample=[]
for i in range(10000):
    #tau_true = tau_true_sample[i]*N+1/(k_0*tau_true_sample[i])
    tau_true = tau_true_sample[i]*(N+k_0)
    mu_t_sample.append(np.random.normal(mu_true, np.power(tau_true,-1)))
    


########################## APPROXIMATED #################

while(control):
    summation = np.sum(np.power(x_real,2))+N*(1/tau_n+ mu_n**2)-2*mu_n*np.sum(x_real)
    b = b_0 + k_0/2 *(mu_n**2+1/tau_n+mu_0**2-2*mu_n*mu_0) + 1/2*summation
    
    tau = (k_0+N)*a_n / b
    if abs(tau_n - tau)<1e-5:
        if abs(b-b_n)<1e-5:
            control = False
    b_n = b
    tau_n = tau
    axs[n_iter//3][n_iter%3].set_title("iter n= "+ str(n_iter))
    
    
    sns.kdeplot(np.random.normal(mu_n, np.power(tau_n,-1), 10000), np.power(np.random.gamma(a_n,1/b_n, size = 10000),-1), color= 'r', ax= axs[n_iter//3][n_iter%3], alpha= 0.3) 
    sns.kdeplot(mu_t_sample, tau_true_sample, color= 'blue', ax= axs[n_iter%2][n_iter%3])
    axs[n_iter//3][n_iter%3].plot(mu_true, (a_0+N/2)/(b_0 + 1/2 * np.sum(np.power(x_real-mean,2)) + k_0*N/(2*(k_0+N))*(mean-mu_0)**2), marker= 'x',)
    n_iter +=1
plt.savefig('comparison.png')
```

# Again, following https://arxiv.org/pdf/1310.5438.pdf

```python
a_0 = 1
b_0 = 2
m = [0,0]
# where we have a prior on alpha, and x 
D = 3 #the dimensions of the problem
n = X_t.shape[0] # numero dati 
eps= np.zeros((X_t.shape[0],1))
```

```python
y_t[y_t==0] = -1
```

```python
def lambd(a):
    a = -abs(a)
    return 0.25 * exprel(a) / (np.exp(a) + 1)   #to avoid problems with small a

def sigma(x):
    return 1/(1+np.e**(-x))


# the only quanity we can already implement 
a_n = a_0 + D/2

b_n = 1 #for beginning 
max_iter = 50
stop = False
n_iter = 0 
V_n = np.identity(D)
b_n = 10
w_n = np.zeros((1,D))
while(not stop):
    A = np.zeros((D,D))
    summ = 0
    for i in range(n):
        A += np.transpose(X_t[i:i+1, ::]).dot(X_t[i:i+1, ::])*lambd(eps[i])
        summ += y_t[i]/2*X_t[i:i+1, ::]
    
    V = np.linalg.inv(a_n/b_n * np.identity(D)+ 2* A )
    w = V.dot(np.transpose(summ))
    b = b_0 + 0.5* (np.transpose(w).dot(w) + np.trace(V))
    n_iter +=1
    
    print(np.linalg.norm(V-V_n, 'fro'),np.linalg.norm(w-w_n), np.abs(b-b_n) )
    
    if(np.linalg.norm(V-V_n, 'fro')< 1e-3 and np.linalg.norm(w-w_n)< 1e-3 and np.abs(b-b_n)< 1e-3):
        stop= True
    if (n_iter>max_iter):
        stop = True
    V_n = V
    w_n = w
    b_n = b
    
    H = V_n+w.dot(np.transpose(w))
    
    for i in range(n):
        eps[i] = np.sqrt(X_t[i:i+1, ::].dot(H).dot(np.transpose(X_t[i:i+1, ::])))
```

```python
pred = np.array(sigma(X_test.dot(m_n)))
for i in range(pred.shape[0]):
    if pred[i]>0.5:
        pred[i] = 1
    else:
        pred[i]=-1
     

confusion_matrix = pd.crosstab(y_test, pred.reshape(pred.shape[0],), rownames=['Actual'], colnames=['Predicted'])
confusion_matrix
```

```python
# would like to construct the real posterior for mu
# w - Ber(0.5)
# clutter N(0, 1)
# prior mean (0,2.3)
# N(mu,1)

# P(x|mu) = 0.5 N(mu*z, 1) + 0.5 N(0,1)
# P(mu) = N(mu_0,2.3)
mu_0 = 0
x = []
z_true=[]
size_sample = 300
for i in range(size_sample):
    if np.random.rand()<0.5:
        x.append(np.random.normal(2,1))
        z_true.append(1)
    else:
        x.append(np.random.normal(0,1))
        z_true.append(0)
    
# initial guess 
```

```python
n_iter  = 1e4
mu_n = [1.6]*(int(n_iter)+1)
z_history = []
sigma_0 = 2
z_n = [round(i) for i in list(np.random.random(size = size_sample))]
z_history = np.zeros(( int(n_iter), size_sample))
# now update until convergence 

for i in range(int(n_iter)-1):
    for j in range(size_sample):
        h = np.exp(-0.5*(x[j]-mu_n[i])**2)
        if np.random.rand() < np.power((h+np.exp(-0.5*np.power(x[j],2))),-1)*h:
            z_n[j] = 1
        else:    
            z_n[j] = 0
        
    z_history[i]= z_n
    mu_n[i] = np.random.normal(sigma_0/(sigma_0 + 1/np.sum(z_n))* pd.Series(x)[pd.Series(z_n)==1].mean(), (np.sum(z_n)+1/sigma_0)**(-1))
    
kernel = stat.gaussian_kde(mu_n[5000:])
h = np.linspace(-3,3,1000)
plt.plot(h, kernel(h))
```
